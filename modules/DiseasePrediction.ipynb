````python
# ============================================================
# ULTRA-FAST TRANSFER LEARNING SOLUTION
# ============================================================
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, applications, callbacks
import time

print("TensorFlow version:", tf.__version__)

# Clear session to free memory
keras.backend.clear_session()

# ============================================================
# 1. DATA LOADING (SAME AS BEFORE)
# ============================================================
base_path = "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)"
train_dir = os.path.join(base_path, "train")
valid_dir = os.path.join(base_path, "valid")

# SIMPLE data augmentation
train_datagen = keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.1
)

valid_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

# Use SMALLER batch for speed
BATCH_SIZE = 32  # Keep this reasonable

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(224, 224),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

num_classes = len(train_generator.class_indices)
print(f"Classes: {num_classes}")
print(f"Training images: {train_generator.samples}")
print(f"Validation images: {valid_generator.samples}")

# ============================================================
# 2. USE MOBILENETV2 (LIGHTNING FAST)
# ============================================================
print("\n" + "="*60)
print("BUILDING MOBILENETV2 MODEL")
print("="*60)

# Load pre-trained MobileNetV2 (very fast, small)
base_model = applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,  # Don't include the classifier
    weights='imagenet'  # Pre-trained on ImageNet
)

# Freeze the base model
base_model.trainable = False

# Build the model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),  # Much faster than Flatten
    layers.Dropout(0.2),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ============================================================
# 3. TRAIN WITH LIMITED STEPS (FAST!)
# ============================================================
print("\n" + "="*60)
print("FAST TRAINING - 5 EPOCHS ONLY")
print("="*60)

# Use SMALL number of steps for fast training
LIMITED_STEPS = 100  # Instead of 2196!
VALIDATION_STEPS = 30  # Instead of 549!

callbacks_list = [
    callbacks.ModelCheckpoint('fast_model.h5', save_best_only=True, monitor='val_accuracy'),
    callbacks.EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
]

start_time = time.time()

history = model.fit(
    train_generator,
    steps_per_epoch=LIMITED_STEPS,  # CRITICAL: Limited steps
    epochs=5,  # Just 5 epochs to start
    validation_data=valid_generator,
    validation_steps=VALIDATION_STEPS,  # Limited validation steps
    callbacks=callbacks_list,
    verbose=1
)

training_time = time.time() - start_time
print(f"\n‚è±Ô∏è  Training completed in {training_time:.0f} seconds ({training_time/60:.1f} minutes)")

# ============================================================
# 4. QUICK EVALUATION
# ============================================================
print("\n" + "="*60)
print("QUICK EVALUATION")
print("="*60)

model.load_weights('fast_model.h5')

# Quick validation
results = model.evaluate(valid_generator, steps=20, verbose=1)
print(f"\nüìä Quick Accuracy: {results[1]:.2%}")

# Make predictions
valid_generator.reset()
predictions = model.predict(valid_generator, steps=20, verbose=1)

# Visualize results
plt.figure(figsize=(12, 4))

# Accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train', marker='o')
plt.plot(history.history['val_accuracy'], label='Val', marker='s')
plt.title(f'Accuracy: {results[1]:.2%}')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train', marker='o')
plt.plot(history.history['val_loss'], label='Val', marker='s')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# 5. IF ACCURACY IS GOOD (>50%), CONTINUE TRAINING
# ============================================================
if results[1] > 0.5:
    print("\n" + "="*60)
    print("CONTINUE TRAINING (Accuracy looks good!)")
    print("="*60)

    # Option 1: Train for more epochs
    print("\nOption 1: Train for 5 more epochs")
    history2 = model.fit(
        train_generator,
        steps_per_epoch=LIMITED_STEPS,
        epochs=5,
        validation_data=valid_generator,
        validation_steps=VALIDATION_STEPS,
        verbose=1
    )

    # Final evaluation
    final_results = model.evaluate(valid_generator, steps=50, verbose=1)
    print(f"\nüéØ FINAL ACCURACY: {final_results[1]:.2%}")

    # Save final model
    model.save('final_plant_disease_model.h5')
    print("‚úì Model saved as 'final_plant_disease_model.h5'")

else:
    print("\n‚ö† Accuracy is low. Trying fine-tuning...")

    # Unfreeze some layers for fine-tuning
    base_model.trainable = True

    # Freeze first 100 layers
    for layer in base_model.layers[:100]:
        layer.trainable = False

    # Recompile with lower learning rate
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("Fine-tuning for 3 epochs...")
    history2 = model.fit(
        train_generator,
        steps_per_epoch=50,  # Even fewer steps for fine-tuning
        epochs=3,
        validation_data=valid_generator,
        validation_steps=20,
        verbose=1
    )

    # Save fine-tuned model
    model.save('fine_tuned_model.h5')
    print("‚úì Fine-tuned model saved")

# ============================================================
# BONUS: EVEN FASTER OPTION (TinyML Style)
# ============================================================
print("\n" + "="*60)
print("BONUS: ULTRA-FAST TINY MODEL")
print("="*60)

# If you need even faster training:
def build_tiny_model(num_classes):
    """Extremely small model for quick testing"""
    model = models.Sequential([
        layers.Input(shape=(224, 224, 3)),

        # Tiny conv blocks
        layers.Conv2D(16, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        layers.GlobalAveragePooling2D(),  # Much smaller than Flatten
        layers.Dense(64, activation='relu'),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

print("\nTo try the tiny model, run:")
print("""
tiny_model = build_tiny_model(num_classes)
tiny_model.summary()

# Train super fast
tiny_history = tiny_model.fit(
    train_generator,
    steps_per_epoch=50,  # Very few steps
    epochs=5,
    validation_data=valid_generator,
    validation_steps=20,
    verbose=1
)
""")
````

````python
# ============================================================
# FIXED & OPTIMIZED MODEL EVALUATION
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pandas as pd
import os
import json

print("TensorFlow version:", tf.__version__)

# ============================================================
# 1. LOAD MODEL & DATA (FIXED VERSION)
# ============================================================
print("\n" + "="*60)
print("MODEL EVALUATION - FIXED VERSION")
print("="*60)

# Load model
model = keras.models.load_model('final_plant_disease_model.h5')
print("‚úì Model loaded")

# Load class indices
with open('class_indices.json', 'r') as f:
    class_indices = json.load(f)

class_names = list(class_indices.keys())
print(f"‚úì {len(class_names)} classes loaded")

# ============================================================
# 2. QUICK ACCURACY (ALREADY HAVE THIS: 86.25%)
# ============================================================
print(f"\nüéØ QUICK ACCURACY RESULT: 86.25%")
print("   (From previous evaluation)")

# ============================================================
# 3. EFFICIENT PREDICTIONS (NO MEMORY ISSUES)
# ============================================================
print("\n" + "="*60)
print("EFFICIENT PREDICTIONS ANALYSIS")
print("="*60)

# Setup validation data
base_path = "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)"
valid_dir = os.path.join(base_path, "valid")

valid_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# METHOD 1: Use model.predict directly (most efficient)
print("\nGenerating predictions efficiently...")
valid_generator.reset()

# Get all predictions at once (TensorFlow handles batching)
y_pred_probs = model.predict(valid_generator, verbose=1)

# Get true labels
y_true = valid_generator.classes
y_pred = np.argmax(y_pred_probs, axis=1)

# Calculate accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"\n‚úÖ FINAL VALIDATION ACCURACY: {accuracy:.2%}")
print(f"   Correct: {np.sum(y_true == y_pred):,} / {len(y_true):,}")

# ============================================================
# 4. TOP/BOTTOM PERFORMING CLASSES
# ============================================================
print("\n" + "="*60)
print("CLASS PERFORMANCE ANALYSIS")
print("="*60)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
per_class_accuracy = cm.diagonal() / cm.sum(axis=1)

# Create analysis dataframe
performance_df = pd.DataFrame({
    'Class': class_names,
    'Accuracy': per_class_accuracy,
    'Samples': cm.sum(axis=1)
}).sort_values('Accuracy')

print("\nüî¥ TOP 10 LOWEST ACCURACY CLASSES:")
print("-" * 65)
for idx, row in performance_df.head(10).iterrows():
    class_short = row['Class'][:35] + "..." if len(row['Class']) > 35 else row['Class']
    print(f"  {class_short:38s} {row['Accuracy']:6.1%} ({row['Samples']:4d} samples)")

print("\nüü¢ TOP 10 HIGHEST ACCURACY CLASSES:")
print("-" * 65)
for idx, row in performance_df.tail(10).iterrows():
    class_short = row['Class'][:35] + "..." if len(row['Class']) > 35 else row['Class']
    print(f"  {class_short:38s} {row['Accuracy']:6.1%} ({row['Samples']:4d} samples)")

# ============================================================
# 5. CONFIDENCE ANALYSIS
# ============================================================
print("\n" + "="*60)
print("CONFIDENCE ANALYSIS")
print("="*60)

confidence_scores = np.max(y_pred_probs, axis=1)
correct_mask = (y_true == y_pred)

print(f"\nüìä CONFIDENCE STATISTICS:")
print(f"  Overall average confidence: {np.mean(confidence_scores):.1%}")
print(f"  Correct predictions: {np.mean(confidence_scores[correct_mask]):.1%}")
print(f"  Incorrect predictions: {np.mean(confidence_scores[~correct_mask]):.1%}")

# Quick histogram
plt.figure(figsize=(10, 4))
plt.hist(confidence_scores[correct_mask], bins=30, alpha=0.7, label='Correct', color='green')
plt.hist(confidence_scores[~correct_mask], bins=30, alpha=0.7, label='Incorrect', color='red')
plt.xlabel('Confidence')
plt.ylabel('Count')
plt.title('Prediction Confidence Distribution')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ============================================================
# 6. SAMPLE PREDICTIONS VISUALIZATION
# ============================================================
print("\n" + "="*60)
print("SAMPLE PREDICTIONS")
print("="*60)

# Get a small batch for visualization
valid_generator.reset()
batch_images, batch_labels = next(valid_generator)
batch_preds = model.predict(batch_images, verbose=0)

# Display 9 samples
plt.figure(figsize=(15, 12))
for i in range(min(9, len(batch_images))):
    plt.subplot(3, 3, i+1)
    plt.imshow(batch_images[i])

    true_idx = np.argmax(batch_labels[i])
    pred_idx = np.argmax(batch_preds[i])
    confidence = np.max(batch_preds[i])

    true_class = class_names[true_idx]
    pred_class = class_names[pred_idx]

    # Shorten long names
    true_short = true_class if len(true_class) < 20 else true_class[:17] + "..."
    pred_short = pred_class if len(pred_class) < 20 else pred_class[:17] + "..."

    color = 'green' if true_idx == pred_idx else 'red'
    plt.title(f"True: {true_short}\nPred: {pred_short}\nConf: {confidence:.1%}",
              fontsize=9, color=color)
    plt.axis('off')

plt.suptitle(f'Sample Validation Predictions | Model Accuracy: {accuracy:.2%}', fontsize=16)
plt.tight_layout()
plt.show()

# ============================================================
# 7. TEST IMAGE PREDICTIONS
# ============================================================
print("\n" + "="*60)
print("TEST IMAGE PREDICTIONS")
print("="*60)

test_dir = "/kaggle/input/new-plant-diseases-dataset/test"
if os.path.exists(test_dir):
    test_images = []
    test_filenames = []

    for img_file in sorted(os.listdir(test_dir)):
        if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(test_dir, img_file)
            test_filenames.append(img_file)

            img = keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
            img_array = keras.preprocessing.image.img_to_array(img) / 255.0
            test_images.append(img_array)

    if test_images:
        test_images = np.array(test_images)
        print(f"Found {len(test_images)} test images")

        # Predict
        test_preds = model.predict(test_images, verbose=1)
        test_classes = np.argmax(test_preds, axis=1)
        test_confidences = np.max(test_preds, axis=1)

        # Display first 12
        plt.figure(figsize=(20, 15))
        for i in range(min(12, len(test_images))):
            plt.subplot(3, 4, i+1)
            plt.imshow(test_images[i])

            pred_class = class_names[test_classes[i]]
            confidence = test_confidences[i]
            filename = test_filenames[i]

            # Simple disease detection from filename
            is_healthy = 'Healthy' in filename
            disease_type = 'Healthy' if is_healthy else 'Diseased'

            plt.title(f"{filename[:15]}...\n{pred_class[:20]}\n{confidence:.1%}\n({disease_type})",
                     fontsize=8)
            plt.axis('off')

        plt.suptitle('Test Image Predictions', fontsize=16)
        plt.tight_layout()
        plt.show()

        # Print predictions table
        print(f"\nüìã TEST IMAGE PREDICTIONS TABLE:")
        print("-" * 85)
        print(f"{'Filename':25s} {'Predicted Disease':35s} {'Confidence':12s}")
        print("-" * 85)

        for i in range(len(test_images)):
            filename = test_filenames[i]
            pred_class = class_names[test_classes[i]]
            confidence = test_confidences[i]

            # Shorten if needed
            filename_display = filename[:22] + "..." if len(filename) > 25 else filename
            class_display = pred_class[:32] + "..." if len(pred_class) > 35 else pred_class

            print(f"{filename_display:25s} {class_display:35s} {confidence:11.1%}")
else:
    print("Test directory not found")

# ============================================================
# 8. ERROR ANALYSIS
# ============================================================
print("\n" + "="*60)
print("ERROR ANALYSIS")
print("="*60)

# Find misclassified samples
misclassified_idx = np.where(y_true != y_pred)[0]
misclassified_count = len(misclassified_idx)

print(f"\nTotal misclassified: {misclassified_count:,} ({misclassified_count/len(y_true):.1%})")

if misclassified_count > 0:
    # Analyze most common errors
    error_pairs = []
    for idx in misclassified_idx[:1000]:  # Sample first 1000 errors
        true_class = class_names[y_true[idx]]
        pred_class = class_names[y_pred[idx]]
        error_pairs.append((true_class, pred_class))

    from collections import Counter
    common_errors = Counter(error_pairs).most_common(10)

    print("\nüîç MOST COMMON MISTAKES:")
    print("-" * 70)
    for (true_class, pred_class), count in common_errors:
        true_short = true_class[:25] + "..." if len(true_class) > 28 else true_class
        pred_short = pred_class[:25] + "..." if len(pred_class) > 28 else pred_class
        print(f"  {true_short:28s} ‚Üí {pred_short:28s}: {count:3d} times")

# ============================================================
# 9. PERFORMANCE SUMMARY
# ============================================================
print("\n" + "="*60)
print("PERFORMANCE SUMMARY")
print("="*60)

print(f"\nüìä MODEL PERFORMANCE METRICS:")
print("-" * 40)
print(f"Overall Accuracy:          {accuracy:.2%}")
print(f"Total Validation Samples:  {len(y_true):,}")
print(f"Correct Predictions:       {np.sum(y_true == y_pred):,}")
print(f"Incorrect Predictions:     {np.sum(y_true != y_pred):,}")

# Performance rating
if accuracy >= 0.9:
    rating = "EXCELLENT üéâ - Production ready"
elif accuracy >= 0.85:
    rating = "VERY GOOD üëç - Could be deployed"
elif accuracy >= 0.8:
    rating = "GOOD ‚úÖ - May need minor improvements"
elif accuracy >= 0.7:
    rating = "FAIR ‚ö†Ô∏è - Consider more training"
else:
    rating = "NEEDS WORK üîß - Retrain with adjustments"

print(f"\nüèÜ PERFORMANCE RATING: {rating}")

# Save summary
summary = {
    'accuracy': float(accuracy),
    'total_samples': int(len(y_true)),
    'correct_predictions': int(np.sum(y_true == y_pred)),
    'misclassified': int(np.sum(y_true != y_pred)),
    'performance_rating': rating
}

with open('model_performance_summary.json', 'w') as f:
    json.dump(summary, f, indent=4)

print("\n‚úì Summary saved to 'model_performance_summary.json'")

print(f"\n‚úÖ EVALUATION COMPLETE! Your model accuracy: {accuracy:.2%}")
````

````python
# ============================================================
# FIXED VERSION - UPDATED FOR GRADIO 4.x
# ============================================================
import gradio as gr
import tensorflow as tf
import numpy as np
from PIL import Image
import json
import os

print("="*60)
print("üåø PLANT DISEASE DETECTOR - WORKING VERSION")
print("="*60)

# Load model - using .keras format (modern)
MODEL_PATH = 'plant_disease_detector.keras'  # Use the .keras version
CLASS_FILE = 'class_indices.json'

# Load model
if os.path.exists(MODEL_PATH):
    model = tf.keras.models.load_model(MODEL_PATH)
    print(f"‚úÖ Model loaded from: {MODEL_PATH}")
else:
    # Try .h5 as backup
    MODEL_PATH = 'final_plant_disease_model.h5'
    model = tf.keras.models.load_model(MODEL_PATH)
    print(f"‚úÖ Model loaded from: {MODEL_PATH}")

# Load class names
with open(CLASS_FILE, 'r') as f:
    class_indices = json.load(f)
class_names = {v: k for k, v in class_indices.items()}
print(f"‚úÖ Loaded {len(class_names)} disease classes")

def predict_top_disease(image):
    """Return only the most confident prediction"""
    try:
        if image is None:
            return "‚ùå Please upload an image first"

        # Preprocess image
        img = image.resize((224, 224))
        img_array = np.array(img) / 255.0

        # Handle different image formats
        if len(img_array.shape) == 2:  # Grayscale
            img_array = np.stack([img_array]*3, axis=-1)
        elif img_array.shape[2] == 4:  # RGBA
            img_array = img_array[:, :, :3]

        img_array = np.expand_dims(img_array, axis=0)

        # Make prediction
        predictions = model.predict(img_array, verbose=0)[0]

        # Get ONLY the most confident prediction
        top_idx = np.argmax(predictions)
        confidence = float(predictions[top_idx])
        disease_name = class_names.get(top_idx, f"Unknown Disease")

        # Parse disease name
        if "___" in disease_name:
            plant, condition = disease_name.split("___", 1)
            condition = condition.replace("_", " ")
        else:
            plant, condition = "Plant", disease_name

        # Determine if healthy
        is_healthy = "healthy" in disease_name.lower()

        # Create clean result
        if is_healthy:
            result = f"""
            # üü¢ **HEALTHY PLANT DETECTED**

            ## **{plant}**

            üìä **Confidence:** {confidence:.1%}

            ‚úÖ **Status:** No disease detected. Plant appears healthy.

            ---
            *Model Accuracy: 92.44%*
            """
        else:
            result = f"""
            # üî¥ **DISEASE DETECTED**

            ## **{plant}**
            ### **{condition}**

            üìä **Confidence:** {confidence:.1%}

            ‚ö†Ô∏è **Status:** Disease detected. Immediate attention recommended.

            ---
            *Model Accuracy: 92.44%*
            """

        return result

    except Exception as e:
        return f"‚ùå Error: {str(e)}"

# Create clean interface
with gr.Blocks(title="Plant Disease Detector", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # üåø **Plant Disease Detector**
    ### Upload a leaf photo ‚Üí Get instant diagnosis
    *92.44% Accurate | 38 Disease Classes*
    """)

    with gr.Row():
        with gr.Column(scale=1):
            image_input = gr.Image(
                label="üì∑ **Upload Leaf Image**",
                type="pil",
                height=300
            )

            analyze_btn = gr.Button(
                "üîç **Analyze Disease**",
                variant="primary",
                size="lg"
            )

            gr.Markdown("""
            **üìù Tips for best results:**
            1. Clear photo of single leaf
            2. Good lighting
            3. Avoid blurry images
            """)

        with gr.Column(scale=2):
            # Result output
            result_output = gr.Markdown(
                label="**DETECTION RESULT**",
                value="### üëà Upload an image to get started"
            )

    # Connect button
    analyze_btn.click(
        fn=predict_top_disease,
        inputs=image_input,
        outputs=result_output
    )

    # Also analyze when image is uploaded
    image_input.change(
        fn=predict_top_disease,
        inputs=image_input,
        outputs=result_output
    )

# Launch with correct syntax
print("\nüöÄ Launching interface...")
demo.launch(share=True, debug=False)
````
