```python
# Create a composite fertility score (example: weighted combination)
# We'll normalize each parameter first, then combine
from sklearn.preprocessing import MinMaxScaler

# Select the 9 key soil attributes
soil_features = ['Clay', 'OM', 'CEC', 'pH', 'V', 'exP', 'exK', 'exCa', 'exMg']
X = soil_df[soil_features].copy()

# Normalize features to 0-1 scale
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=soil_features)

# Define weights for each parameter (adjust based on crop importance)
# Higher weight = more important for fertility
weights = {
    'OM': 0.15,      # Organic Matter - very important
    'pH': 0.15,      # pH - critical for nutrient availability
    'exP': 0.13,     # Phosphorus
    'exK': 0.13,     # Potassium
    'CEC': 0.12,     # Cation Exchange Capacity
    'exCa': 0.10,    # Calcium
    'exMg': 0.10,    # Magnesium
    'V': 0.07,       # Base Saturation
    'Clay': 0.05     # Clay content
}

# Calculate weighted fertility score
soil_df['Fertility_Score'] = 0
for feature, weight in weights.items():
    soil_df['Fertility_Score'] += X_scaled[feature] * weight

# Classify into Fertile/Not Fertile based on threshold
threshold = soil_df['Fertility_Score'].median()  # Using median as initial threshold
soil_df['Fertility_Label'] = np.where(soil_df['Fertility_Score'] >= threshold, 'Fertile', 'Not_Fertile')

print("‚úÖ Fertility label created!")
print(f"Threshold score: {threshold:.3f}")
print("\nüìä Fertility distribution:")
print(soil_df['Fertility_Label'].value_counts())
print(f"\nüìà Score range: {soil_df['Fertility_Score'].min():.3f} to {soil_df['Fertility_Score'].max():.3f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, hamming_loss, classification_report, f1_score
import numpy as np

# Fix: Use stratified sampling for multi-label data (preserve label distributions)
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit

# Install the library if needed
!pip install iterative-stratification -q

# Create stratified train-test split
msss = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=0.25, random_state=42)
for train_idx, test_idx in msss.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fix: Use balanced class weights and adjust for crops with 0 positives
adjusted_estimators = []
for i, crop in enumerate(y.columns):
    # Check if we have both classes
    class_counts = y_train[crop].value_counts()

    if len(class_counts) == 1:
        print(f"‚ö†Ô∏è  Warning: {crop} has only one class ({class_counts.index[0]}) in training data")
        # For crops with only one class, we'll use a dummy classifier
        from sklearn.dummy import DummyClassifier
        dummy_clf = DummyClassifier(strategy="constant", constant=class_counts.index[0])
        adjusted_estimators.append(dummy_clf)
    else:
        # Use RandomForest with balanced class_weight
        rf = RandomForestClassifier(
            n_estimators=150,
            random_state=42,
            class_weight='balanced',
            min_samples_split=5,  # Helps with small classes
            min_samples_leaf=2
        )
        adjusted_estimators.append(rf)

# Create custom multi-output classifier
class AdjustedMultiOutputClassifier:
    def __init__(self, estimators):
        self.estimators = estimators

    def fit(self, X, y):
        for i, estimator in enumerate(self.estimators):
            estimator.fit(X, y.iloc[:, i])
        return self

    def predict(self, X):
        predictions = []
        for estimator in self.estimators:
            pred = estimator.predict(X)
            predictions.append(pred)
        return np.column_stack(predictions)

    def predict_proba(self, X):
        probabilities = []
        for estimator in self.estimators:
            # Handle classifiers that might only have one class
            try:
                proba = estimator.predict_proba(X)
                # If binary classification, take probability of positive class
                if proba.shape[1] == 2:
                    probabilities.append(proba[:, 1])
                else:
                    # If only one class, predict 0 or 1 with confidence
                    probabilities.append(proba[:, 0] * 0 + estimator.predict(X))
            except (AttributeError, IndexError):
                # For dummy classifiers or other issues
                pred = estimator.predict(X)
                probabilities.append(pred.astype(float))
        return np.column_stack(probabilities)

# Create and train the model
print("üîÑ Training adjusted multi-label model...")
model = AdjustedMultiOutputClassifier(adjusted_estimators)
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)

# Fix: Calculate F1-score instead of accuracy (better for imbalanced data)
print("\n‚úÖ Improved Model Evaluation:")
print(f"Exact Match Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print(f"Hamming Loss (lower is better): {hamming_loss(y_test, y_pred):.4f}")

print("\nüìã Per-Crop F1-Scores (Better Metric for Imbalance):")
crop_names = y.columns.tolist()
for i, crop in enumerate(crop_names):
    crop_f1 = f1_score(y_test.iloc[:, i], y_pred[:, i], zero_division=0)
    print(f"{crop}: F1-Score = {crop_f1:.3f}")

# Check which crops have predictions
print(f"\nüìä Positive predictions in test set:")
for i, crop in enumerate(crop_names):
    pos_preds = y_pred[:, i].sum()
    pos_actual = y_test.iloc[:, i].sum()
    print(f"{crop}: Predicted {pos_preds} positive, Actually {pos_actual} positive")
```

```python
def predict_crop_suitability_improved(model, scaler, sample_features, crop_names, threshold=0.3):
    """
    Improved prediction function that handles single-class crops.

    Parameters:
    model: Trained AdjustedMultiOutputClassifier
    scaler: Fitted StandardScaler
    sample_features: Array or DataFrame with 9 soil features
    crop_names: List of crop names
    threshold: Confidence threshold for positive prediction

    Returns:
    DataFrame with suitability predictions
    """
    # Scale the sample
    sample_scaled = scaler.transform([sample_features])

    # Get predictions and probabilities
    predictions = model.predict(sample_scaled)[0]
    probabilities = model.predict_proba(sample_scaled)[0]

    # For crops where we only have predictions (no probabilities),
    # use the prediction as probability
    confidences = np.where(
        (probabilities == 0) | (probabilities == 1),
        predictions.astype(float),  # Use 0 or 1 directly
        probabilities  # Use actual probabilities
    )

    # Apply threshold (only for crops with probabilities)
    adjusted_predictions = np.where(
        (confidences > 0) & (confidences < 1),  # Has real probabilities
        (confidences >= threshold).astype(int),  # Apply threshold
        predictions  # Keep original 0/1 prediction
    )

    # Create results
    results = pd.DataFrame({
        'Crop': crop_names,
        'Suitable': adjusted_predictions,
        'Confidence': confidences
    })

    # Sort by confidence (highest first)
    results = results.sort_values('Confidence', ascending=False)

    return results

# Test with the same sample
print("üå± Fixed Prediction for Sample #0:")
sample = X.iloc[0].values

predictions_df = predict_crop_suitability_improved(
    model, scaler, sample, y.columns.tolist(), threshold=0.3
)

print("\nüìä All Crop Predictions with Confidence:")
print(predictions_df)

print("\n‚úÖ Recommended Crops (Confidence > 0.5):")
print(predictions_df[predictions_df['Confidence'] > 0.5][['Crop', 'Suitable', 'Confidence']])
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load your crop list (all 92 crops you mentioned earlier)
all_crops = ['Arhar/Tur', 'Bajra', 'Castor seed', 'Cotton(lint)', 'Dry chillies',
             'Groundnut', 'Horse-gram', 'Jowar', 'Korra', 'Maize', 'Moong(Green Gram)',
             'Other Kharif pulses', 'Ragi', 'Rice', 'Sugarcane', 'Sunflower', 'Tobacco',
             'Gram', 'Wheat', 'Samai', 'Sesamum', 'Urad', 'Small millets', 'Varagu',
             'Mesta', 'Soyabean', 'Rapeseed &Mustard', 'Oilseeds total', 'Jute',
             'Linseed', 'Other Rabi pulses', 'Arecanut', 'Niger seed', 'Onion',
             'Potato', 'Sweet potato', 'Tapioca', 'Turmeric', 'Dry ginger', 'Barley',
             'Banana', 'Coriander', 'Garlic', 'Sannhamp', 'Safflower', 'Moth',
             'Pulses total', 'Paddy', 'Black pepper', 'Cardamom', 'Cashewnut',
             'Total foodgrain', 'Kapas', 'Peas & beans (Pulses)', 'Masoor', 'Khesari',
             'Blackgram', 'Other Cereals & Millets', 'other oilseeds', 'Cond-spcs other',
             'Tea', 'Guar seed', 'Jute & mesta', 'other misc. pulses', 'Rubber',
             'Coffee', 'Beans & Mutter(Vegetable)', 'Bhindi', 'Brinjal', 'Citrus Fruit',
             'Cucumber', 'Grapes', 'Mango', 'Orange', 'other fibres', 'Other Fresh Fruits',
             'Other Vegetables', 'Papaya', 'Pome Fruit', 'Tomato', 'Cabbage',
             'Peas (vegetable)', 'Bottle Gourd', 'Pineapple', 'Turnip', 'Carrot', 'Redish',
             'Arcanut (Processed)', 'Atcanut (Raw)', 'Cashewnut Processed',
             'Cashewnut Raw', 'Bitter Gourd', 'Drum Stick', 'Jack Fruit', 'Snak Guard',
             'Cauliflower', 'Water Melon', 'Ash Gourd', 'Beet Root', 'Lab-Lab',
             'Other Citrus Fruit', 'Pome Granet', 'Ribed Guard', 'Yam', 'Pump Kin', 'Apple',
             'Peach', 'Pear', 'Plums', 'Ber', 'Litchi', 'Other Dry Fruit', 'Cowpea(Lobia)',
             'Ginger']

print(f"üìä Total crops to analyze: {len(all_crops)}")

# Define crop categories with IDEAL conditions (not minimum requirements)
crop_categories = {
    'acid_loving': {
        'crops': ['Rice', 'Potato', 'Blueberry', 'Tea', 'Coffee', 'Pineapple', 'Sweet potato'],
        'ideal_pH': (5.0, 5.8),
        'importance': {'pH': 0.4, 'OM': 0.3, 'exP': 0.15, 'exK': 0.15}
    },
    'moderately_acidic': {
        'crops': ['Maize', 'Tomato', 'Cabbage', 'Carrot', 'Onion', 'Peas', 'Beans'],
        'ideal_pH': (5.5, 6.5),
        'importance': {'pH': 0.3, 'exCa': 0.2, 'OM': 0.2, 'exP': 0.15, 'exK': 0.15}
    },
    'nutrient_demanding': {
        'crops': ['Sugarcane', 'Banana', 'Tapioca', 'Turmeric', 'Ginger'],
        'ideal_pH': (5.5, 7.0),
        'importance': {'exK': 0.3, 'exP': 0.25, 'OM': 0.25, 'pH': 0.2}
    },
    'drought_tolerant': {
        'crops': ['Sorghum', 'Millet', 'Groundnut', 'Sunflower', 'Castor'],
        'ideal_pH': (5.5, 7.5),
        'importance': {'Clay': 0.3, 'CEC': 0.3, 'exK': 0.2, 'pH': 0.2}
    },
    'legumes': {
        'crops': ['Soybean', 'Peas', 'Beans', 'Lentils', 'Groundnut', 'Chickpea'],
        'ideal_pH': (6.0, 7.0),
        'importance': {'pH': 0.4, 'exCa': 0.25, 'exP': 0.2, 'OM': 0.15}
    }
}

# Calculate suitability score for a single crop
def calculate_crop_score(soil_sample, crop_name, crop_type):
    """Calculate how suitable a soil is for a specific crop (0-100 scale)"""
    score = 0

    # 1. pH suitability (bell curve - best at ideal range)
    ideal_pH = crop_type['ideal_pH']
    pH = soil_sample['pH']

    if ideal_pH[0] <= pH <= ideal_pH[1]:
        # Perfect match - center of range
        pH_score = 100
    elif pH < ideal_pH[0]:
        # Too acidic - linear penalty
        pH_score = max(0, 100 - (ideal_pH[0] - pH) * 30)
    else:
        # Too alkaline
        pH_score = max(0, 100 - (pH - ideal_pH[1]) * 30)

    score += pH_score * crop_type['importance'].get('pH', 0)

    # 2. Nutrient scores (more is better, but with diminishing returns)
    for nutrient, weight in crop_type['importance'].items():
        if nutrient == 'pH':
            continue

        if nutrient in soil_sample:
            # Normalize nutrient value (0-100 scale)
            # Using your dataset's distribution
            if nutrient == 'OM':
                # Organic Matter: 14-37 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 14) / (37-14) * 100)
            elif nutrient == 'exP':
                # Phosphorus: 4-104 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 4) / (104-4) * 100)
            elif nutrient == 'exK':
                # Potassium: 0.9-10.3 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 0.9) / (10.3-0.9) * 100)
            elif nutrient == 'exCa':
                # Calcium: 8-78 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 8) / (78-8) * 100)
            elif nutrient == 'Clay':
                # Clay: 175-511 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 175) / (511-175) * 100)
            elif nutrient == 'CEC':
                # CEC: 37.5-148.9 in your data
                nutrient_score = min(100, (soil_sample[nutrient] - 37.5) / (148.9-37.5) * 100)
            else:
                nutrient_score = 50  # Default

            score += nutrient_score * weight

    return min(100, max(0, score))  # Clamp to 0-100

# Create recommendation system
def recommend_crops_for_soil(soil_sample_df, top_n=10):
    """Recommend best crops for a given soil sample"""

    # Ensure soil_sample is a dictionary of the relevant features
    if isinstance(soil_sample_df, pd.DataFrame):
        soil_sample = soil_sample_df.iloc[0].to_dict()
    elif isinstance(soil_sample_df, pd.Series):
        soil_sample = soil_sample_df.to_dict()
    else:
        # If it's already a dictionary or another compatible type, use it directly
        soil_sample = soil_sample_df

    recommendations = []

    # First, check which crop category each crop belongs to
    crop_to_category = {}
    for category, info in crop_categories.items():
        for crop in info['crops']:
            crop_to_category[crop] = category

    # Score all crops
    for crop in all_crops:
        # Get crop category (default to moderately_acidic for unknown crops)
        category = crop_to_category.get(crop, 'moderately_acidic')
        crop_type = crop_categories.get(category, crop_categories['moderately_acidic'])

        score = calculate_crop_score(soil_sample, crop, crop_type)

        # Only include crops with reasonable scores
        if score >= 30:  # Minimum threshold
            recommendations.append({
                'Crop': crop,
                'Category': category,
                'Score': round(score, 1),
                'Suitability': 'Excellent' if score >= 80 else
                              'Good' if score >= 60 else
                              'Fair' if score >= 40 else 'Marginal'
            })

    # Sort by score and return top N
    recommendations_df = pd.DataFrame(recommendations)
    if not recommendations_df.empty:
        recommendations_df = recommendations_df.sort_values('Score', ascending=False).head(top_n)

    return recommendations_df

# Test with your first soil sample
print("üå± Crop Recommendations for Sample #0:")
sample_0 = soil_df.iloc[0]
recommendations = recommend_crops_for_soil(sample_0, top_n=15)

print(recommendations.to_string(index=False))
print(f"\nüìä Your soil pH: {sample_0['pH']:.1f}")
print(f"üìä Best suited category: Acid-loving crops (pH 5.0-5.8)")
```

```python
!pip install gradio plotly -q

import gradio as gr
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# ======================
# 1. SOIL HEALTH ANALYSIS
# ======================

def analyze_soil_health(Clay, OM, CEC, pH, V, exP, exK, exCa, exMg):
    """Comprehensive soil health analysis with ratings"""

    # Calculate individual parameter scores (0-100)
    scores = {}
    ratings = {}
    recommendations = []

    # 1. pH Analysis (most critical)
    if 6.0 <= pH <= 7.0:
        scores['pH'] = 100
        ratings['pH'] = "‚úÖ Excellent"
        recommendations.append("pH is in optimal range for most crops")
    elif 5.5 <= pH < 6.0 or 7.0 < pH <= 7.5:
        scores['pH'] = 70
        ratings['pH'] = "‚ö†Ô∏è Moderate"
        if pH < 6.0:
            recommendations.append("Soil is slightly acidic. Consider small lime application")
        else:
            recommendations.append("Soil is slightly alkaline. May need sulfur amendment")
    else:
        scores['pH'] = 30
        ratings['pH'] = "‚ùå Poor"
        if pH < 5.5:
            recommendations.append("Strongly acidic soil. Lime application required for most crops")
        else:
            recommendations.append("Strongly alkaline soil. Sulfur or gypsum needed")

    # 2. Organic Matter Analysis
    if OM >= 30:
        scores['OM'] = 100
        ratings['OM'] = "‚úÖ Excellent"
        recommendations.append("High organic matter - good for soil structure and nutrients")
    elif 20 <= OM < 30:
        scores['OM'] = 70
        ratings['OM'] = "‚ö†Ô∏è Moderate"
        recommendations.append("Organic matter is adequate but could be improved with compost")
    else:
        scores['OM'] = 40
        ratings['OM'] = "‚ùå Low"
        recommendations.append("Low organic matter. Add compost, manure, or green manure crops")

    # 3. Nutrient Analysis (N-P-K-Ca-Mg)
    nutrient_status = []

    # Phosphorus
    if exP >= 20:
        scores['P'] = 100
        ratings['P'] = "‚úÖ Sufficient"
    elif 10 <= exP < 20:
        scores['P'] = 65
        ratings['P'] = "‚ö†Ô∏è Marginal"
        nutrient_status.append("Phosphorus")
    else:
        scores['P'] = 30
        ratings['P'] = "‚ùå Deficient"
        nutrient_status.append("Phosphorus")

    # Potassium
    if exK >= 2.5:
        scores['K'] = 100
        ratings['K'] = "‚úÖ Sufficient"
    elif 1.5 <= exK < 2.5:
        scores['K'] = 65
        ratings['K'] = "‚ö†Ô∏è Marginal"
        nutrient_status.append("Potassium")
    else:
        scores['K'] = 30
        ratings['K'] = "‚ùå Deficient"
        nutrient_status.append("Potassium")

    # Calcium
    if exCa >= 40:
        scores['Ca'] = 100
        ratings['Ca'] = "‚úÖ Sufficient"
    elif 20 <= exCa < 40:
        scores['Ca'] = 65
        ratings['Ca'] = "‚ö†Ô∏è Marginal"
        nutrient_status.append("Calcium")
    else:
        scores['Ca'] = 30
        ratings['Ca'] = "‚ùå Deficient"
        nutrient_status.append("Calcium")

    # Magnesium
    if exMg >= 20:
        scores['Mg'] = 100
        ratings['Mg'] = "‚úÖ Sufficient"
    elif 10 <= exMg < 20:
        scores['Mg'] = 65
        ratings['Mg'] = "‚ö†Ô∏è Marginal"
    else:
        scores['Mg'] = 30
        ratings['Mg'] = "‚ùå Deficient"

    # 4. CEC and Base Saturation
    if CEC >= 80:
        scores['CEC'] = 100
        ratings['CEC'] = "‚úÖ High"
        recommendations.append("High CEC - good nutrient holding capacity")
    elif 50 <= CEC < 80:
        scores['CEC'] = 75
        ratings['CEC'] = "‚ö†Ô∏è Moderate"
    else:
        scores['CEC'] = 40
        ratings['CEC'] = "‚ùå Low"
        recommendations.append("Low CEC - may need frequent fertilization")

    # Base Saturation
    if V >= 80:
        scores['V'] = 100
        ratings['V'] = "‚úÖ High"
    elif 60 <= V < 80:
        scores['V'] = 70
        ratings['V'] = "‚ö†Ô∏è Adequate"
    else:
        scores['V'] = 40
        ratings['V'] = "‚ùå Low"
        recommendations.append("Low base saturation - soil may be acidic")

    # 5. Overall Soil Health Score
    weights = {
        'pH': 0.25,    # Most important
        'OM': 0.20,    # Very important
        'P': 0.15,     # Important nutrients
        'K': 0.15,
        'Ca': 0.10,
        'CEC': 0.10,
        'V': 0.05
    }

    overall_score = sum(scores[param] * weights.get(param, 0) for param in weights)

    if overall_score >= 80:
        health_rating = "Excellent"
        health_color = "green"
    elif overall_score >= 60:
        health_rating = "Good"
        health_color = "orange"
    elif overall_score >= 40:
        health_rating = "Fair"
        health_color = "yellow"
    else:
        health_rating = "Poor"
        health_color = "red"

    # 6. Create summary
    summary = {
        'overall_score': round(overall_score, 1),
        'health_rating': health_rating,
        'health_color': health_color,
        'ratings': ratings,
        'scores': scores,
        'recommendations': recommendations,
        'nutrient_deficiencies': nutrient_status
    }

    return summary

# ======================
# 2. CROP RECOMMENDATIONS
# ======================

def get_crop_recommendations(soil_summary, soil_params):
    """Get crop recommendations based on soil health analysis"""

    # Crop database with requirements
    crops_database = {
        # Acid-loving crops
        'Rice': {'min_pH': 4.5, 'max_pH': 6.5, 'min_OM': 15, 'min_P': 10, 'min_K': 1.5, 'category': 'Staple'},
        'Potato': {'min_pH': 4.8, 'max_pH': 5.5, 'min_OM': 20, 'min_P': 25, 'min_K': 3.0, 'category': 'Vegetable'},
        'Tea': {'min_pH': 4.5, 'max_pH': 5.5, 'min_OM': 25, 'min_P': 15, 'min_K': 2.0, 'category': 'Cash'},
        'Coffee': {'min_pH': 4.5, 'max_pH': 6.0, 'min_OM': 30, 'min_P': 15, 'min_K': 2.5, 'category': 'Cash'},
        'Pineapple': {'min_pH': 4.5, 'max_pH': 5.5, 'min_OM': 20, 'min_P': 15, 'min_K': 2.0, 'category': 'Fruit'},

        # Moderately acidic to neutral
        'Maize': {'min_pH': 5.5, 'max_pH': 7.0, 'min_OM': 20, 'min_P': 15, 'min_K': 2.0, 'category': 'Staple'},
        'Tomato': {'min_pH': 5.5, 'max_pH': 6.8, 'min_OM': 25, 'min_P': 20, 'min_K': 2.5, 'min_Ca': 30, 'category': 'Vegetable'},
        'Wheat': {'min_pH': 6.0, 'max_pH': 7.5, 'min_OM': 25, 'min_P': 20, 'min_K': 2.5, 'min_Ca': 30, 'category': 'Staple'},
        'Soybean': {'min_pH': 6.0, 'max_pH': 7.0, 'min_OM': 20, 'min_P': 15, 'min_K': 2.0, 'category': 'Legume'},

        # Nutrient demanding
        'Sugarcane': {'min_pH': 5.5, 'max_pH': 7.5, 'min_OM': 30, 'min_P': 25, 'min_K': 3.0, 'category': 'Cash'},
        'Banana': {'min_pH': 5.5, 'max_pH': 7.0, 'min_OM': 25, 'min_P': 20, 'min_K': 3.0, 'category': 'Fruit'},

        # Drought tolerant
        'Groundnut': {'min_pH': 5.5, 'max_pH': 6.5, 'min_OM': 15, 'min_P': 12, 'min_K': 1.5, 'category': 'Legume'},
        'Castor': {'min_pH': 5.5, 'max_pH': 7.0, 'min_OM': 15, 'min_P': 10, 'min_K': 1.5, 'category': 'Oilseed'},
    }

    recommendations = []

    for crop, requirements in crops_database.items():
        suitability = 100

        # Check pH suitability
        if not (requirements['min_pH'] <= soil_params['pH'] <= requirements['max_pH']):
            suitability -= 40

        # Check organic matter
        if soil_params['OM'] < requirements['min_OM']:
            suitability -= 30

        # Check nutrients
        if 'min_P' in requirements and soil_params['exP'] < requirements['min_P']:
            suitability -= 20
        if 'min_K' in requirements and soil_params['exK'] < requirements['min_K']:
            suitability -= 15
        if 'min_Ca' in requirements and soil_params['exCa'] < requirements['min_Ca']:
            suitability -= 10

        # Only include suitable crops
        if suitability >= 50:
            recommendations.append({
                'Crop': crop,
                'Suitability': suitability,
                'Category': requirements['category'],
                'Rating': 'Excellent' if suitability >= 85 else
                         'Good' if suitability >= 70 else
                         'Fair' if suitability >= 60 else 'Marginal'
            })

    # Sort by suitability
    recommendations.sort(key=lambda x: x['Suitability'], reverse=True)

    return recommendations[:10]  # Return top 10

# ======================
# 3. VISUALIZATIONS
# ======================

def create_soil_health_radar(soil_summary):
    """Create radar chart for soil health"""

    # Parameters to show
    params = ['pH', 'OM', 'P', 'K', 'Ca', 'CEC']
    values = [soil_summary['scores'].get(param, 50) for param in params]

    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=params,
        fill='toself',
        name='Soil Health',
        line_color='blue',
        fillcolor='rgba(135, 206, 250, 0.5)'
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(
                visible=True,
                range=[0, 100],
                tickfont=dict(size=10)
            ),
            angularaxis=dict(tickfont=dict(size=11))
        ),
        showlegend=False,
        height=350,
        margin=dict(l=50, r=50, t=50, b=50)
    )

    return fig

def create_nutrient_bars(soil_params):
    """Create bar chart for nutrient levels"""

    nutrients = ['P (mg/kg)', 'K (mg/kg)', 'Ca (mg/kg)', 'Mg (mg/kg)']
    values = [soil_params['exP'], soil_params['exK'], soil_params['exCa'], soil_params['exMg']]

    # Define optimal ranges
    optimal_min = [15, 2.0, 30, 15]
    optimal_max = [40, 4.0, 60, 30]

    fig = go.Figure()

    # Add bars for current values
    fig.add_trace(go.Bar(
        x=nutrients,
        y=values,
        name='Current Level',
        marker_color=['#FF6B6B' if v < optimal_min[i] else '#4ECDC4' if v <= optimal_max[i] else '#FFD166' for i, v in enumerate(values)]
    ))

    # Add optimal range indicators
    for i, nutrient in enumerate(nutrients):
        fig.add_shape(
            type="rect",
            xref="x",
            yref="y",
            x0=i-0.4,
            x1=i+0.4,
            y0=optimal_min[i],
            y1=optimal_max[i],
            fillcolor="rgba(78, 205, 196, 0.2)",
            line_width=0,
            layer="below"
        )

    fig.update_layout(
        title="Nutrient Levels vs Optimal Range",
        yaxis_title="Concentration",
        height=300,
        showlegend=False,
        margin=dict(l=50, r=50, t=50, b=50)
    )

    return fig

# ======================
# 4. MAIN GRADIO INTERFACE
# ======================

def analyze_and_recommend(Clay, OM, CEC, pH, V, exP, exK, exCa, exMg):
    """Main function that analyzes soil and provides recommendations"""

    # Create soil parameters dict
    soil_params = {
        'Clay': Clay, 'OM': OM, 'CEC': CEC, 'pH': pH, 'V': V,
        'exP': exP, 'exK': exK, 'exCa': exCa, 'exMg': exMg
    }

    # Step 1: Analyze soil health
    soil_summary = analyze_soil_health(Clay, OM, CEC, pH, V, exP, exK, exCa, exMg)

    # Step 2: Get crop recommendations
    crop_recommendations = get_crop_recommendations(soil_summary, soil_params)

    # Step 3: Create HTML outputs
    # --- Soil Health Report ---
    health_html = f"""
    <div style='background-color: #f8f9fa; padding: 20px; border-radius: 10px; border-left: 5px solid {soil_summary["health_color"]};'>
        <h2 style='color: {soil_summary["health_color"]}; margin-top: 0;'>üå± Soil Health: {soil_summary["health_rating"]}</h2>
        <div style='font-size: 24px; font-weight: bold; color: {soil_summary["health_color"]};'>
            Overall Score: {soil_summary["overall_score"]}/100
        </div>
    </div>

    <h3>üìä Parameter Analysis</h3>
    <table style='width: 100%; border-collapse: collapse;'>
        <tr style='background-color: #e9ecef;'>
            <th style='padding: 10px; text-align: left;'>Parameter</th>
            <th style='padding: 10px; text-align: left;'>Rating</th>
            <th style='padding: 10px; text-align: left;'>Score</th>
        </tr>
    """

    for param, rating in soil_summary['ratings'].items():
        if param in soil_summary['scores']:
            health_html += f"""
            <tr style='border-bottom: 1px solid #dee2e6;'>
                <td style='padding: 10px;'><strong>{param}</strong></td>
                <td style='padding: 10px;'>{rating}</td>
                <td style='padding: 10px;'>{soil_summary['scores'][param]}/100</td>
            </tr>
            """

    health_html += "</table>"

    # --- Recommendations ---
    health_html += f"""
    <h3>üí° Recommendations</h3>
    <div style='background-color: #fff3cd; padding: 15px; border-radius: 5px; border-left: 4px solid #ffc107;'>
    """

    for rec in soil_summary['recommendations'][:5]:  # Show top 5 recommendations
        health_html += f"<p>‚Ä¢ {rec}</p>"

    if soil_summary['nutrient_deficiencies']:
        health_html += f"<p><strong>Nutrient Deficiencies:</strong> {', '.join(soil_summary['nutrient_deficiencies'])}</p>"

    health_html += "</div>"

    # --- Crop Recommendations ---
    crops_html = "<h2>üåæ Recommended Crops</h2>"

    if crop_recommendations:
        crops_html += """
        <table style='width: 100%; border-collapse: collapse; margin-top: 20px;'>
            <tr style='background-color: #28a745; color: white;'>
                <th style='padding: 12px; text-align: left;'>Rank</th>
                <th style='padding: 12px; text-align: left;'>Crop</th>
                <th style='padding: 12px; text-align: left;'>Category</th>
                <th style='padding: 12px; text-align: left;'>Suitability</th>
                <th style='padding: 12px; text-align: left;'>Rating</th>
            </tr>
        """

        for i, crop in enumerate(crop_recommendations[:8], 1):
            # Color based on rating
            color_map = {
                'Excellent': '#28a745',
                'Good': '#ffc107',
                'Fair': '#fd7e14',
                'Marginal': '#dc3545'
            }

            crops_html += f"""
            <tr style='border-bottom: 1px solid #dee2e6;'>
                <td style='padding: 10px;'>{i}</td>
                <td style='padding: 10px; font-weight: bold;'>{crop['Crop']}</td>
                <td style='padding: 10px;'>{crop['Category']}</td>
                <td style='padding: 10px;'>{crop['Suitability']}/100</td>
                <td style='padding: 10px; color: {color_map.get(crop['Rating'], '#000')}; font-weight: bold;'>{crop['Rating']}</td>
            </tr>
            """

        crops_html += "</table>"

        # Additional note
        crops_html += f"""
        <div style='margin-top: 20px; padding: 15px; background-color: #d4edda; border-radius: 5px; border-left: 4px solid #28a745;'>
            <p><strong>üí° Note:</strong> Based on your soil's pH of {pH} and overall health score of {soil_summary['overall_score']},
            these crops are most likely to thrive. Acid-loving crops (like Rice, Potato, Tea) perform best in your soil conditions.</p>
        </div>
        """
    else:
        crops_html += """
        <div style='padding: 20px; background-color: #f8d7da; border-radius: 5px; border-left: 4px solid #dc3545;'>
            <p><strong>‚ö†Ô∏è No suitable crops found.</strong> Your soil may require significant amendments before planting.</p>
            <p>Consider addressing the issues identified in the soil health analysis first.</p>
        </div>
        """

    # Return all outputs
    return (
        health_html,  # Soil health analysis
        crops_html,   # Crop recommendations
        create_soil_health_radar(soil_summary),  # Radar chart
        create_nutrient_bars(soil_params)        # Nutrient chart
    )

# ======================
# 5. BUILD GRADIO UI
# ======================

with gr.Blocks(theme=gr.themes.Soft(), title="Soil Health & Crop Advisor") as demo:

    # Header
    gr.Markdown("# üå± Soil Health Analysis & Crop Recommendation System")
    gr.Markdown("Analyze your soil health first, then get personalized crop recommendations based on the analysis.")

    with gr.Row():
        # Left Column: Inputs
        with gr.Column(scale=1):
            gr.Markdown("## üìù Soil Parameters Input")

            with gr.Group():
                gr.Markdown("### Physical Properties")
                Clay = gr.Slider(100, 600, value=350, step=10, label="Clay Content (g/kg)")
                OM = gr.Slider(10, 50, value=25, step=1, label="Organic Matter (g/kg)")
                CEC = gr.Slider(30, 150, value=80, step=5, label="CEC (cmol‚Çä/kg)")

            with gr.Group():
                gr.Markdown("### Chemical Properties")
                pH = gr.Slider(4.0, 8.0, value=5.5, step=0.1, label="pH (4.0 = Acidic, 7.0 = Neutral, 8.0 = Alkaline)")
                V = gr.Slider(20, 100, value=65, step=5, label="Base Saturation V (%)")

            with gr.Group():
                gr.Markdown("### Nutrient Levels")
                with gr.Row():
                    exP = gr.Slider(5, 100, value=20, step=5, label="Available P (mg/kg)")
                    exK = gr.Slider(0.5, 15, value=3.0, step=0.5, label="Available K (mg/kg)")
                with gr.Row():
                    exCa = gr.Slider(10, 100, value=35, step=5, label="Available Ca (mg/kg)")
                    exMg = gr.Slider(5, 60, value=18, step=2, label="Available Mg (mg/kg)")

            # Example buttons
            gr.Markdown("### üìã Example Soil Profiles")
            with gr.Row():
                gr.Examples(
                    examples=[
                        [406, 32, 63, 5.3, 76, 15, 4.1, 31, 13],  # Your acidic sample
                        [250, 35, 95, 6.5, 85, 35, 4.5, 50, 25],  # Good all-around soil
                        [400, 15, 45, 4.8, 50, 8, 1.2, 15, 8]     # Poor soil needing amendments
                    ],
                    inputs=[Clay, OM, CEC, pH, V, exP, exK, exCa, exMg],
                    label="Click to load examples"
                )

            analyze_btn = gr.Button("üîç Analyze Soil & Get Recommendations", variant="primary", size="lg")

        # Right Column: Outputs
        with gr.Column(scale=2):
            gr.Markdown("## üìä Analysis Results")

            # Tabs for organized output
            with gr.Tabs():
                with gr.TabItem("1Ô∏è‚É£ Soil Health Analysis"):
                    soil_health_html = gr.HTML(label="Soil Health Report")

                with gr.TabItem("2Ô∏è‚É£ Recommended Crops"):
                    crop_recommendations_html = gr.HTML(label="Crop Recommendations")

                with gr.TabItem("üìà Visualizations"):
                    with gr.Row():
                        radar_plot = gr.Plot(label="Soil Health Radar Chart")
                        nutrient_plot = gr.Plot(label="Nutrient Levels")

    # Footer
    gr.Markdown("---")
    gr.Markdown("""
    **How to use this tool:**
    1. Enter your soil test results in the left panel
    2. Click "Analyze Soil & Get Recommendations"
    3. View soil health analysis in Tab 1
    4. See recommended crops in Tab 2
    5. Check visualizations in Tab 3

    *Note: Recommendations are based on general crop requirements. Local conditions may vary.*
    """)

    # Connect button to function
    analyze_btn.click(
        fn=analyze_and_recommend,
        inputs=[Clay, OM, CEC, pH, V, exP, exK, exCa, exMg],
        outputs=[soil_health_html, crop_recommendations_html, radar_plot, nutrient_plot]
    )

# ======================
# 6. LAUNCH THE APP
# ======================
print("üöÄ Launching Soil Health & Crop Advisor...")
print("üìä Features:")
print("  1. Comprehensive soil health analysis")
print("  2. Individual parameter ratings")
print("  3. Personalized recommendations")
print("  4. Crop suitability scoring")
print("  5. Interactive visualizations")

demo.launch(share=True, debug=False)
```
